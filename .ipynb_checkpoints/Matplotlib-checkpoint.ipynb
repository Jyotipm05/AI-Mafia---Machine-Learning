in a variety of hardcopy formats and interactive environments across platforms.
Matplotlib can be used in Python seripts, the Python and IPython shell, the jupyter
notebook, web application servers, and four graphical user interface toolkits.

## Example 1

Plot a simple line.

- Take a graph paper! (import matplotlib)


```python
import matplotlib.pyplot as plt
```

* Specify data to plot

|x||1|2|3||---||---|---|---||y|2|4|1|

```python
x = [1, 2, 3]
y = [2, 4, 1]
```

* Plot points on graph paper!

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

* Specify X and Y axis labels

```python
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.title('codingclub')
plt.show()
```
