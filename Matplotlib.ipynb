## Example 7

Plotting pie-chart

```python
# defining labels
activities = ["eat", "sleep", "work", "play"]
```

```python
# portion covered by each label
slices = [3, 7, 8, 6]
```

```python
# color for each label
colors = ["r", "m", "g", "b"]
```

```python
# plotting the pie chart
plt.pie(
    slices,
    labels=activities,
    colors=colors,
    startangle=90,
    shadow=True,
    explode=(0, 0, 0.1, 0),
    radius=1.2,
    autopct="%1.1f%%",
)
```

```python
plt.close()
```

## Subplot

Subplots are required when we want to show or more plots in same figure. We can do it in two ways using two slightyly different methods.

## Example 9

Plotting subplot using method 1

```python
# function to regeneate coordiantes
def create_plot(ptype):
    # setting the x-axis values
    x = np.arange(-10, 10, 0.01)
    # setting the y-axis values
    if ptype == "linear":
        y = x
    elif ptype == "quadratic":
        y = x**2
    elif ptype == "cubic":
        y = x**3
    elif ptype == "quartic":
        y = x**4
    return (x, y)
```

Figure acts as a top level container for all plot elements.

```python
# create a figure
fig = plt.figure()
```

```python
# define subplots and their positions in figure
plt1 = fig.add_subplot(221)
plt2 = fig.add_subplot(222)
plt3 = fig.add_subplot(223)
plt4 = fig.add_subplot(224)
```
