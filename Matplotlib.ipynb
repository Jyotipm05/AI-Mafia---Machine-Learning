## Example 1

Plot a simple line.

- Take a graph paper! (import matplotlib)


```python
import matplotlib.pyplot as plt
```

* Specify data to plot

|x||1|2|3||---||---|---|---||y|2|4|1|

```python
x = [1, 2, 3]
y = [2, 4, 1]
```

* Plot points on graph paper!

```python
plt.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

* Specify X and Y axis labels

```python
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
plt.title('codingclub')
plt.show()
```

* Give title to your graph

```python
plt.title("My first garph!")
```

* Show it

```python
plt.show()
```

```python
# close the graph
plt.close()
```

## Example 2 &para;

Plot 2 lines on same plot

```python
# line 1 points
x1 = [1, 2, 3]
y1 = [2, 4, 1]

# line 2 points
x2 = [1, 2, 3]
y2 = [4, 1, 3]
```

```python
# plot 1
plt.plot(x1, y1, label="line 1")
# plot 2
plt.plot(x2, y2, label="line 2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
plt.close()
```

## Example 3

Customizing line plot

```python
plt.style.available
```

```python
plt.style.use('fivethirtyeight')
```

```python
x = [2, 2, 3, 4, 5, 6]
y = [2, 4, 3, 5, 2, 6]
```

```python
plt.plot(x,y,color='green',linestyle='dashed',linewidth=3,marker='o',markerfacecolor="blue",markersize=8)
```

```python
plt.xlim(1, 8)
plt.ylim(1, 8)
```

```python
plt.show()
```

```python
plt.close()
```

## Example 4

Plotting bar chart.

```python
# x-coordinates of left sides of bars
left = [1, 2, 3, 4, 5]
```

```python
# heights of
height = [10, 24, 36, 40, 5]
```

```python
# labels for bars
tick_label = ["one", "two", "three", "four", "five"]
```

```python
# plotting a bar chart
plt.bar(left, height, tick_label=tick_label, width=0.8, color=["red", "blue"])
```

```python
plt.close()
```

## Example 5

Plot histogram

```python
ages = [
    2,
    5,
    70,
    40,
    30,
    45,
    50,
    45,
    43,
    40,
    44,
    60,
    7,
    13,
    57,
    18,
    90,
    77,
    32,
    21,
    20,
    40,
]
```

```python
# setting the ranges and no. o fintervals
range = (0, 100)
bins = 10
```

```python
# plotting a histogram
plt.hist(ages, bins, range, color="red", histtype="bar", rwidth=0.8)
plt.xlabel("age")
plt.ylabel("No. of people")
```

```python
plt.close()
```
